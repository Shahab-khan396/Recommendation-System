```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, 
# {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

# we want to extract only the specific terms from the genres and store it in a list
# [Action, Adventure, Fantasy, Science Fiction] extracting only these terms from the above string and store it in a list


def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

# with the help of above code we can convert the string into a list of dictionaries and then we can extract the name of the genres and store it in a list
```

```python
movies['genres']=movies['genres'].apply(convert) 
```
